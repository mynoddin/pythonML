```python
import matplotlib.pyplot as plt
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 123, 10]

# Create a line plot
plt.plot(x, y, label='y = 2x')  # Line plot with label
plt.xlabel('X-axis')            # Label for X-axis
plt.ylabel('Y-axis')            # Label for Y-axis
plt.title('Line Plot Example')  # Title of the plot
plt.legend()                    # Add legend
plt.show()                      # Display the plot
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.scatter(x, y, color='red', marker='o')  # Scatter plot
plt.title('Scatter Plot Example')
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values, color='blue')  # Bar plot
plt.title('Bar Plot Example')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Example: Load Iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target

# Pairplot to visualize relationships
sns.pairplot(data, hue='species')
plt.show()
```

```python
epochs = [1, 2, 3, 4, 5]
train_loss = [0.8, 0.6, 0.4, 0.3, 0.2]
val_loss = [0.9, 0.7, 0.5, 0.4, 0.35]

plt.plot(epochs, train_loss, label='Training Loss')
plt.plot(epochs, val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc

# Generate synthetic data
X, y = make_classification(n_samples=1000, n_features=20, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# Train a model
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_prob = clf.predict_proba(X_test)[:, 1]

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], 'r--')  # Baseline
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
